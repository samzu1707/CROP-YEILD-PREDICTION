<div class="alert alert-block alert-danger">
   <b>Data Pre-Processing</b>
</div>

```python
df = df.drop('Area', axis=1)
```

```python
df = df.drop('Production', axis=1)
```

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
for col in ['Crop', 'Season', 'State']:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
le_crop = LabelEncoder()
le_season = LabelEncoder()
le_state = LabelEncoder()
    
df['Crop'] = le_crop.fit_transform(df['Crop'])
df['Season'] = le_season.fit_transform(df['Season'])
df['State'] = le_state.fit_transform(df['State'])
```

<div class="alert alert-block alert-info">
   <b>Dealing with Null Values</b>
</div>

```python
print(df.isnull().sum())
print("Null values have been handled. You should see all zeros above.")

print(df.dtypes)
print("Categorical columns have been converted to int-like types by Label Encoding.")
```

<div class="alert alert-block alert-warning">
   <b>Changing Datatypes of Columns to be "Int"</b>
</div>

```python
for col in df.select_dtypes('object').columns:
    df[col] = pd.factorize(df[col])[0]

print("Final Data Types after conversion:")
print(df.dtypes)
```

<div class="alert alert-block alert-success">
   <b>Train test Splitting</b>
</div>

```python
X = df.iloc[:, :-1] 
Y = df.iloc[:, -1]  

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, shuffle=False, random_state=42)

print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"Y_train shape: {Y_train.shape}")
print(f"Y_test shape: {Y_test.shape}")
```

<div class="alert alert-block alert-danger">
   <b>Choose and apply the right model</b>
</div>

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)

model.fit(X_train, Y_train)

print("Random Forest Regressor Model Trained Successfully.")
```

<div class="alert alert-block alert-info">
   <b>Testing/Predicting</b>
</div>

```python
Y_pred = model.predict(X_test)

print("Predictions completed.")
```

<div class="alert alert-block alert-warning">
   <b>Display Accuracy Score</b>
</div>

```python
mae = mean_absolute_error(Y_test, Y_pred)
rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
r2 = r2_score(Y_test, Y_pred)
accuracy = r2 * 100

print(f"MAE: {mae:.4f}, RMSE: {rmse:.4f}, R2: {r2:.4f}")
print(f"Approximate Model Accuracy: {accuracy:.2f}%")
```

```python
joblib.dump(model, 'random forest regressor model.pkl')
print("Model saved as 'random forest regressor model.pkl'")
```
